```python
benigns = pd.DataFrame(
    {
        'id': [],
        "digit": [],
        'path': []
    }
)
benigns = benigns.set_index('id')
lambdas = pd.DataFrame(
    {
        'id': [],
        'digit': [],
        'metric': [],
        'lambda': [],
        'path': []
    }
)
lambdas = lambdas.set_index('id')
```

```python
for i, (rd, dirs, files) in enumerate(os.walk(root)):
    if i == 0:
        continue
    digit = rd[-1]
    for file in files:
        if 'benign' in file:
            id = len(benigns.index)
            benigns.loc[id] = [digit, os.path.join(rd, file)[12:]]
        else:
            file_splits = file.split('_')
            id = len(lambdas.index)
            lambdas.loc[id] = [
                
                digit,
                file_splits[2],
                file_splits[3],
                os.path.join(rd, file)[12:]
            ]

    
```

```python
with open('tex_helper.txt', 'w', encoding='UTF8') as out:
    for metric in lambdas.metric.unique():
        out.write(f'% {metric}\n')
        out.write('\\begin{figure}\n')
        for digit in sorted(benigns.digit.unique()):
            out.write('\t\\begin{subfigure}[b]{\\textwidth}\n')
            out.write('\t\t\\includegraphics[width=0.1\\textwidth]{'\
                      +f'{benigns[benigns.digit == digit].path.array[0]}'.replace('\\', '/')+'}\n')
            for lambd in ['0.001', '0.01', '0.1', '1', '10', '100', '1000']:
                out.write('\t\t\\includegraphics[width=0.1\\textwidth]{'\
                      +f'{lambdas[lambdas.metric == metric][lambdas.digit == digit][lambdas["lambda"] == f"lambda={lambd}"].path.array[0]}'.replace('\\', '/')+'}\n')
            out.write('\t\\end{subfigure}\n')

        out.write('\t\\centering\n\t\\caption{Vzorky generované CW útokem za použití metriky '
                + f'{metric}' \
                +'}\n\t\\label{cw_' \
                + f'{metric}'\
                +'}\n\\end{figure}\n\n')
```
